En este caso podemos ver que hay una baja correlación entre las variables.

```python
# Verificar si hay correlación entre los datos
# 1. Filtramos el DataFrame para quedarnos solo con las columnas numéricas
df_numeric = df.select_dtypes(include=np.number)
# 2. Calculamos la matriz de correlación
corr = df_numeric.corr()
# Opcional: puedes imprimir la matriz para ver los valores
print(corr)
# 3. Graficamos la matriz de correlación
# El gráfico mostrará la relación entre area, geoname_id, numeric y population
sm.graphics.plot_corr(corr, xnames=list(corr.columns))
plt.show()
```

```python
# Visualizamos
df_espanol.set_index('alpha_3')[['population','area']].plot(kind = 'bar',rot = 65,figsize=(20,10))
plt.show()
```

> Vamos a proceder a hacer `detección de outliers (valores atipicos)` en este caso definimos el limite superior (e inferior) la media mas (menos) "2 veces" la desviación estándar" que muchas veces se toma de maximo de tolerancia.

```python
anomalies = []
# Función para detectar outliers
def find_anomalies(data):
    data_std = data.std()
    data_mean = data.mean()
    anomaly_cut_off = data_std * 2
    lower_limit = data_mean - anomaly_cut_off
    upper_limit = data_mean + anomaly_cut_off
    print(lower_limit.iloc[0])
    print(upper_limit.iloc[0])

    # Generar lista de outliers
    for index,row in data.iterrows():
        outlier = row # Obtener primera columna
        if (outlier.iloc[0] > upper_limit.iloc[0]) or (outlier.iloc[0] < lower_limit.iloc[0]):
            anomalies.append(index)
    return anomalies


find_anomalies(df_espanol.set_index('alpha_3')[['population']])
```

Se detecto como `outlier` a Brasil y a USA. Los eliminamos y graficamos ordenado por poblacion de menor a mayor.

```python
# Quitamos BRA y USA por ser outliers y volvamos a graficar.
df_espanol.drop([30,233],inplace=True)
df_espanol.set_index('alpha_3')[['population','area']].sort_values(by='population').plot(kind = 'bar',rot = 65,figsize=(20,10))
plt.show()
```
